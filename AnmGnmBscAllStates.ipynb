```python
import os
import pandas as pd
folder_path = 'downloads/NRTS_PassbooksData/PhaseII'  # Replace with the actual path to your folder
l = []
# Iterate over files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        # Perform operations on the file
        l.append(pd.read_excel(file_path))
        print(file_path)
```

```python
dfFinal = pd.concat(l, axis = 0)
```

```python
dfFinal['name_of_the_programme_anm_gnm_bsc'].value_counts()
```

```python
ANM_index= dfFinal['name_of_the_programme_anm_gnm_bsc']== 'B.Sc (Nursing)'
```

```python
ANM_index.sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax = dfFinal[ANM_index]['statecode'].value_counts().plot(kind ='bar')

plt.title('B.Scs State-wise')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.show()
```
